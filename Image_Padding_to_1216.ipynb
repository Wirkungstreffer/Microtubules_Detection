```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
os.environ['DISPLAY'] = ':1'

img1 = cv2.imread('images/001_img.png')
reflect = cv2.copyMakeBorder(img1,8,8,8,8,cv2.BORDER_REFLECT)
imgsize = reflect.shape
print(imgsize)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
```
